### 시각화 실습 (1)

```python
# matplotlib 폰트 설정 -> 기본적으로 한글 지원 x
from matplotlib import font_manager, rc
font_path = "data/THEdog.ttf"   #폰트파일의 위치
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# 1
data = {'1분기': [500,690,1100,1500,1990,1020],
       '2분기': [450,700,1030,1650,2020,1600],
       '3분기': [520,820,1200,1700,2300,2200],
       '4분기': [610,900,1380,1850,2420,2550]}
df = pd.DataFrame(data, index=[2015,2016,2017,2018,2019,2020])
display(df)
```

```python
# 2
df.index = df.index.map(lambda x : str(x)+'년') # map 함수 사용하여 index 수정!
plt.figure(figsize=(10,6)) # 그래프 사이즈
plt.plot(df)
plt.title('시각화 과제 1', color='purple') # 제목, 컬러
plt.legend(labels=df) # 범례
plt.savefig("output/hw1.png") # 파일 저장
plt.show()
```

```python
# 3
df2 = df.T
plt.figure(figsize=(10,6))
plt.plot(df2)
plt.legend(labels=df2)
plt.savefig("output/hw2.png")
plt.show()
```

```python
# 4
plt.figure(figsize=(10,6)) # 그리기 전에!!
plt.pie(df.sum(), labels = df.columns, 
        colors = ['red', 'green', 'brown', 'purple' ], autopct='%.1f%%', shadow = True, explode = [0.05,0.05,0.05,0.05])
plt.savefig('output/hw3.png')
plt.show()
       
```

```python
# 5 
plt.figure(figsize=(10,6))
plt.pie(df2.sum(), labels = df2.columns, autopct='%.1f%%', colors = ['red', 'green', 'brown', 'purple', 'yellow', 'pink'])
plt.savefig('output/hw4.png')
plt.show()
```
